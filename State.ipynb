1. What is hypothesis testing in statistics?

-> Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data. It helps you test assumptions or claims about a population parameter.







2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

-> Null Hypothesis:

* It is the starting assumption in hypothesis testing.

* It states that there is no effect, no difference, or no relationship.

* It is assumed to be true unless evidence proves otherwise.


* Example: "The new drug has no effect on blood pressure."

-> Alternative Hypothesis:

* It is the opposite of the null hypothesis.

* It states that there is an effect, a difference, or a relationship.

* This is what you are trying to find evidence in favor of.


* Example: "The new drug does affect blood pressure.





3. What is the significance level in hypothesis testing, and why is it important?

-> Definition: The significance level is the probability of rejecting the null hypothesis when it is actually true. This is also known as a Type l error.

-> Why is it important:

1. Sets the standard for decision making: It tells you how much risk you are willing to take in making a wrong decision.

2. Controls type l error: A lower α reduces the chance of a fa;se positive.

3. Basis for comparison with p-value: After performing the test, you compare the p-value to α.

 4. What does a P-value represent in hypothesis testing?

 -> The P-va;ue measures the strength of evidence against the null hypothesis.

* A small P-value means strong evidence against H₀.
* A large P-value means weak evidence against H₀.



5. How do you interpret the P-value in hypothesis testing?

* Suggests that H₀ is probablly false.

-> A large P-value (like 0.30 or 0.50):

10. What is the T-distribution, and when should it be used instead of the normal distribution?

-> The T-distribution is a type of probability distribution that is used for hypothesis testing and confidence intervals when the sample size is small and the poppulation standard deviation is unknown.

-> When should you use the T-distribution instead of the normal distribution?

* Small sample size: When the sample size is small, the T-distribution accounts for the greater variability in smaller samples.

* Unknown population standard deviation: If you do not know the population standard deviation, and you are using the sample standard deviation to estimate it, you should use the T-distribution. The sample standard deviation introduces more uncerainty, which is why the T-distribution has heavier tails to account for that.

* Normality Assumption: The T-distribution assumes that the data is roughlly normally distributed. If the sample size is large, the central limit theorem ensures that the sampling distribution of the mean will approximate normality, and a Z-test can be used instead of a T-test.






11. What is the difference between a Z-test and a T-test?

-> The Z-test is used when the sample size is large and the population standard deviation is known, as the sampling distribution approximates normality. The T-test, however, is used for small sample size when the population standard deviation is unknown, relying on the sample standard deviation and the T-distribution, which has heavier tails to account for greater variability in small samples. The Z-test uses the population standard deviation and does not require degrees of freedom, while the T-test uses the sample standard deviation and requires degrees of freedom. In short, use the Z-test for large samples with known population parameters and the T-test for small samples pr when population parameters are unknown.

```python
def bernoulli_pdf(x, p):
  """
  Calculate the PDF of a Bernoullli distribution.

  Parameters:
  x (int): The outcome (0 or 1).
  p (float): The probability of success (0 ≤ p ≤ 1).

  Returns:
  float: Probability of the given outcome.
  """
  if x not in [0, 1]:
    raise ValueError("x must be 0 and 1")
  if not (0 <= p <= 1):
    raise ValueError("pmust be between 0 and 1")

  return p**x* (1 - p)**(1 - x)

print("p(X=0):", bernoulli_pdf(0, 0.3))
print("p(X=1):", bernoulli_pdf(1, 0.3))
```

9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
  """
  Calculates Z-scores for a dataset

  Parameters:
  data (list or np.array): Input data.

  Returns:
  np.array:Z-scores.
  """

  mean = np.mean(data)
  std_dev = np.std(data)
  z_scores = (data - mean) / std_dev
  return z_scores

data = np.random.normal(50, 10, 100)

z_scores = calculate_z_scores(data)

plt.plot(z_scores, marker='o', linestyle='', color='purple', alpha=0.7)
plt.axhline(0, color='gray', linestyle='--')
plt.title('Z-scores of the Dataset')
plt.xlabel('Index')
plt.ylabel('Z-score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()


```

19. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

sample = np.array([170, 172, 168, 165, 174, 169, 171, 173, 167, 166])

population_mean = 168
population_std = 3

sample_mean = np.mean(sample)
n = len(sample)

z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z_stat:.2f}")
print(f"P-value = {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
  print("\nRehect the nul hypothesis: Significant difference found.")
else:
  print("\nFail to reject the null hypothesis: No significant difference.")

```

20. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats

data = np.array([52, 55, 50, 53, 54, 56, 51, 52, 55, 53])

mean = np.mean(data)
std_dev = np.std(data, ddof=1)
n = len(data)
confidence_level =0.95
alpha = 1 - confidence_level

t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
margin_of_error = t_critical * (std_dev / np.sqrt(n))

lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Sample Mean = {mean:.2f}")
print(f"Confidence Interval ({int(confidence_level*100)}%): ({lower_bound:.2f})")
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)
    margin = stats.t.ppf((1 + confidence) / 2.0, n - 1) * std_err
    lower = mean - margin
    upper = mean + margin
    return mean, lower, upper


np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)
mean, ci_lower, ci_upper = confidence_interval(sample_data)

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_err = stats.sem(data)
    t_critical = stats.t.ppf((1 + confidence) / 2.0, df=n-1)
    moe = t_critical * std_err
    return moe

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=8, size=40)
moe = margin_of_error(sample_data)

print(f"Sample Size: {len(sample_data)}")
print(f"Confidence Level: 95%")
print(f"Margin of Error: {moe:.2f}")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayesian_update(prior, sensitivity, specificity):
    """
    Applies Bayes' Theorem to compute posterior probability.

    Parameters:
    - prior: P(D), prior probability of having the disease
    - sensitivity: P(Pos | D), test true positive rate
    - specificity: P(Neg | ~D), test true negative rate

    Returns:
    - posterior: P(D | Pos), probability of having the disease given a positive test
    """
    false_positive_rate = 1 - specificity


    evidence = sensitivity * prior + false_positive_rate * (1 - prior)

    posterior = (sensitivity * prior) / evidence
    return posterior


prior = 0.01
sensitivity = 0.90
specificity = 0.95

posterior = bayesian_update(prior, sensitivity, specificity)

print("Bayesian Inference Using Bayes’ Theorem")
print("---------------------------------------")
print(f"Prior Probability (P(D)): {prior:.2%}")
print(f"Sensitivity (P(Pos | D)): {sensitivity:.2%}")
print(f"Specificity (P(Neg | ~D)): {specificity:.2%}")
print(f"Posterior Probability (P(D | Pos)): {posterior:.2%}")
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):

    observed = np.array(observed)

    row_totals = observed.sum(axis=1).reshape(-1, 1)
    col_totals = observed.sum(axis=0).reshape(1, -1)
    grand_total = observed.sum()

    expected = (row_totals @ col_totals) / grand_total

    return expected


observed_data = [
    [10, 20, 30],
    [6,  9,  15]
]

expected = calculate_expected_frequencies(observed_data)

observed_df = pd.DataFrame(observed_data, columns=["A", "B", "C"], index=["Group 1", "Group 2"])
expected_df = pd.DataFrame(expected, columns=["A", "B", "C"], index=["Group 1", "Group 2"])

print("Observed Frequencies:\n", observed_df)
print("\nExpected Frequencies:\n", expected_df.round(2))
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
from scipy.stats import chi2


np.random.seed(42)

sample = np.random.normal(loc=50, scale=10, size=30)

hypothesized_variance = 100

n = len(sample)
sample_variance = np.var(sample, ddof=1)

chi_square_statistic = (n - 1) * sample_variance / hypothesized_variance
print(f"Chi-square Statistic: {chi_square_statistic:.4f}")

df = n - 1


p_value = 2 * min(chi2.cdf(chi_square_statistic, df), 1 - chi2.cdf(chi_square_statistic, df))


print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized variance.")
else:
    print("Fail to reject the null hypothesis: The population variance is not significantly different from the hypothesized variance.")
```

25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
from scipy.stats import norm

x1 = 100
n1 = 200

x2 = 120
n2 = 180

p1_hat = x1 / n1
p2_hat = x2 / n2


p_pool = (x1 + x2) / (n1 + n2)


z_statistic = (p1_hat - p2_hat) / np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")
```

26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the result.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

np.random.seed(42)

sample_1 = np.random.normal(loc=50, scale=10, size=100)

sample_2 = np.random.normal(loc=50, scale=15, size=100)

var_1 = np.var(sample_1, ddof=1)
var_2 = np.var(sample_2, ddof=1)

F_statistic = var_1 / var_2 if var_1 >= var_2 else var_2 / var_1

df1 = len(sample_1) - 1
df2 = len(sample_2) - 1

p_value = 1 - f.cdf(F_statistic, df1, df2)

print(f"F-statistic: {F_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")

x = np.linspace(0, 5, 500)
y = f.pdf(x, df1, df2)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="F-distribution", color="blue")
plt.axvline(F_statistic, color="red", linestyle="--", label=f"F-statistic: {F_statistic:.4f}")
plt.fill_between(x, y, where=(x >= F_statistic), color="red", alpha=0.3, label="Critical region")

plt.title(f"F-Test for Variance Comparison\np-value: {p_value:.4f}", fontsize=16)
plt.xlabel("F-statistic", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt


np.random.seed(42)
observed_data = np.random.choice([1, 2, 3, 4, 5, 6], size=100, replace=True)

expected_frequencies = [100 / 6] * 6

observed_frequencies = [np.sum(observed_data == i) for i in range(1, 7)]

chi2_stat, p_value = chisquare(observed_frequencies, expected_frequencies)

print(f"Observed Frequencies: {observed_frequencies}")
print(f"Expected Frequencies: {expected_frequencies}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not follow the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data follows the expected distribution.")

categories = [1, 2, 3, 4, 5, 6]

plt.figure(figsize=(10, 6))
plt.bar(categories, observed_frequencies, width=0.4, label='Observed', align='center', color='blue', alpha=0.6)
plt.bar(categories, expected_frequencies, width=0.4, label='Expected', align='edge', color='red', alpha=0.6)

plt.title('Chi-square Goodness-of-Fit Test: Observed vs Expected Frequencies', fontsize=14)
plt.xlabel('Die Face', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
